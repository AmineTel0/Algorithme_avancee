```python
#Exercice 2
#1)

def est_suffixe(chp, ch):
    return ch.endswith(chp)

def sigma(P, x):
    m = len(P)
    for k in range(min(m, len(x)), -1, -1):
        if est_suffixe(P[:k], x):
            return k
    return 0

def calcul_fonction_transition(P, alphabet):
    m = len(P)
    delta = [{} for _ in range(m + 1)]
    for q in range(m + 1):
        for a in alphabet:
            delta[q][a] = sigma(P, P[:q] + a)
    return delta

def recherche_automate_fini(T, P, delta):
    q = 0
    m = len(P)
    occ = 0
    for c in T:
        q = delta[q][c]
        if q == m:
            occ += 1
    return occ
```

```python
#2)

import random
import time

def recherche_naive(T, P):
    n = len(T)
    m = len(P)
    occ = 0
    for i in range(n - m + 1):
        if T[i:i+m] == P:
            occ += 1
    return occ

alphabet = ['a', 'b', 'c']

m = int(input("Longueur du motif : "))
n = int(input("Longueur du texte : "))
P = input("Motif : ")

T = ''.join(random.choice(alphabet) for _ in range(n))
print(T[:100])

t1 = time.time()
occ_naive = recherche_naive(T, P)
t2 = time.time()

delta = calcul_fonction_transition(P, alphabet)
t3 = time.time()
occ_auto = recherche_automate_fini(T, P, delta)
t4 = time.time()

print(occ_naive)
print(occ_auto)
print(t2 - t1)
print(t4 - t3)
```
